```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("train.csv")

# View first 5 rows
df.head()

# Basic info
df.info()

# Summary stats
df.describe()

# Column names
df.columns
```

```python
df.isnull().sum()
```

```python
df['Sex'].value_counts()
df['Embarked'].value_counts()
```

```python
df['Sex'].value_counts()
df['Embarked'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Age', 'Fare', 'Pclass']], hue='Survived')
plt.show()
```

```python
print("""
Summary:
1. Younger passengers had slightly higher survival rates.
2. Females survived more than males.
3. Higher ticket fare was associated with higher survival.
4. Passengers in 1st class had higher survival rates than 3rd class.
""")
```
